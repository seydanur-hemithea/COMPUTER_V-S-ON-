```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from easydict import EasyDict as edict
from tqdm import tqdm
from torchvision import transforms as trans

from drive.MyDrive.arcface_project.Yuz_Tanima_Proje_Dosyalari.Kodlar.ArcFace.ArcFace.ArcFace.data.ms1m import get_train_loader
from drive.MyDrive.arcface_project.Yuz_Tanima_Proje_Dosyalari.Kodlar.ArcFace.ArcFace.ArcFace.data.lfw import LFW
from drive.MyDrive.arcface_project.Yuz_Tanima_Proje_Dosyalari.Kodlar.ArcFace.ArcFace.ArcFace.backbone.arcfacenet import SEResNet_IR
from drive.MyDrive.arcface_project.Yuz_Tanima_Proje_Dosyalari.Kodlar.ArcFace.ArcFace.ArcFace.margin.ArcMarginProduct import ArcMarginProduct
from drive.MyDrive.arcface_project.Yuz_Tanima_Proje_Dosyalari.Kodlar.ArcFace.ArcFace.ArcFace.util.utils import save_checkpoint,test
```

```python
conf = edict()

# 📁data paths
conf.train_root = "/content/drive/MyDrive/arcface_project/Yuz_Tanima_Proje_Dosyalari/Kodlar/ArcFace/ArcFace/ArcFace/dataset/MS1M"
conf.lfw_test_root = "/content/drive/MyDrive/arcface_project/Yuz_Tanima_Proje_Dosyalari/Kodlar/ArcFace/ArcFace/ArcFace/dataset/lfw_aligned_112"
conf.lfw_file_list="/content/drive/MyDrive/arcface_project/Yuz_Tanima_Proje_Dosyalari/Kodlar/ArcFace/ArcFace/ArcFace/dataset/lfw_pair.txt"

conf.mode='se_ir'#'ir'
conf.depth=50
conf.margin_type='ArcFace'
conf.feature_dim=512
conf.scale_siz=32.0


# 🧠TRAİNİNG SETTİNGS
conf.batch_size = 96# if you decrrease batch_size you should decrease learning rate(lr)
conf.lr = 0.01# you should decrease epoch by epoch lr to train loss is low
conf.milestones=[8,10,12]#The learning rate will be decayed which epochs.
conf.total_epoch=14

conf.num_workers = 4
conf.embedding_size = 512
conf.pin_memory=True

# 🧱 Model SETTİNGS
conf.backbone = "SEResNet_IR"
conf.margin_type = "ArcMarginProduct"
conf.device = "cuda"  # "cpu" olarak da ayarlanabilir

# 📦SAVE AND LOG
conf.log_path = "/content/drive/MyDrive/arcface_project/Yuz_Tanima_Proje_Dosyalari/Logs"
conf.save_path = "/content/drive/MyDrive/arcface_project/Yuz_Tanima_Proje_Dosyalari/Checkpoints"
conf.save_dir=os.path.join(conf.save_path,conf.mode+'_'+str(conf.depth))
```

```python
transform=trans.Compose([
    trans.ToTensor(),#scaling
    trans.Normalize(mean=(0.5,0.5,0.5),std=(0.5,0.5,0.5))

])#many transform get together,
trainloader,class_num=get_train_loader(conf)
```

```python
print("number of id :", class_num)
```

```python

print(trainloader.dataset)
```

```python
#The procedures above are also performed on the test dataset.

lfwdataset = LFW(conf.lfw_test_root, conf.lfw_file_list, transform = transform)
#In essence, the operation applies the transform function to the labeled images found in conf.lfw_file_list, which are stored under the directory defined by conf.lfw_test_root.

#A DataLoader is required to iterate over the incoming dataset.
lfwloader = torch.utils.data.DataLoader(lfwdataset, batch_size = 128, num_workers = conf.num_workers)
```

```python
net = SEResNet_IR(conf.depth, feature_dim = conf.feature_dim, mode = conf.mode).to(conf.device) #model
margin = ArcMarginProduct(conf.feature_dim, class_num).to(conf.device) #ArcFace use, ArcFace applies loss function .
```

```python

print(net)
```

```python

criterion = nn.CrossEntropyLoss()#loss function
```

```python

optimizer = optim.SGD([#model parameters
    {"params": net.parameters(), "weight_decay":5e-4}, #What we did above is essentially this: we applied the transform transformation to the labeled images listed in conf.lfw_file_list, located at the path specified by conf.lfw_test_root.
    {"params": margin.parameters(), "weight_decay":5e-4}
], lr = conf.lr, momentum = 0.9, nesterov = True)#lerarning rate,update gradians
```

```python
print(optimizer)
```

```python

def schedule_lr():#this function provide update lr
    for params in optimizer.param_groups:
        params["lr"]/=10
    print(optimizer)
```

```python
best_acc = 0
for epoch in range(1, conf.total_epoch+1):
    net.train()#dropout runing
    #net.eval dropout not using

    print("epoch {}/{}".format(epoch, conf.total_epoch), flush = True)

    if epoch == conf.milestones[0]: #8.epoch
        schedule_lr()
    if epoch == conf.milestones[1]: #10.epoch
        schedule_lr()
    if epoch == conf.milestones[2]: #12.epoch
        schedule_lr()
# befrore than again this codes run  you should run model and optimizer cells
    for data in tqdm(trainloader): #for visualization tqdm
        img, label = data[0].to(conf.device), data[1].to(conf.device)
        optimizer.zero_grad() #optimizer gradyans get zeros for confusing during Backward elimination brings errors

        logits = net(img) #We send the images to the model and retrieve the output.


# (96,512) vektor
        output = margin(logits,label)
        total_loss = criterion(output, label)
        total_loss.backward()
        optimizer.step()

    #test
    net.eval()
    lfw_acc = test(conf, net,lfwdataset, lfwloader)
    print("\nLFW: {:.4f} | train_loss: {:.4f}\n".format(lfw_acc, total_loss.item()))
    is_best = lfw_acc > best_acc
    best_acc = max(lfw_acc, best_acc)
    save_checkpoint({
        "epoch":epoch,
        "net_state_dict": net.state_dict(),
        "margin_state_dict":margin.state_dict(),
        "best_acc": best_acc
    }, is_best,checkpoint = conf.save_dir)
```
